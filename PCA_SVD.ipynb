## SVD

```python
u,sigma,vt = np.linalg.svd(image1)
```

```python
Sigma = np.zeros((1024,1024))
```

### Rank 0

```python
i0 = u@Sigma@vt
```

```python
plt.imshow(i0)
```

### Rank 1

```python
Sigma[0][0] = sigma[0]
i1 = u@Sigma@vt

plt.imshow(i1, cmap='Greys_r')
```

### Rank 10

```python
for i in range(1,10):
    Sigma[i][i] = sigma[i]
```

```python
i10 = u@Sigma@vt
plt.imshow(i10, cmap='Greys_r')
```

```python
class ImageCompressor:
    def __init__(self, image_name=None):
        ######################################
        
        self.image_name = image_name
        
        from matplotlib.image import imread
        
        
        ######################################
        
        if self.image_name is not None:
            self.img_arr = imread(image_name)
            self.red = self.img_arr[:,:,0]
            self.green = self.img_arr[:,:,1]
            self.blue = self.img_arr[:,:,2]
        else:
            print('Cannot Create Class. No Image name supplied.')
    
    def compress(self, pca_comps=25):
        '''This function accepts number of Principal Components as an estimate of size. By default, this is set to 25, change
           if necessary.'''
        
        import numpy as np
        from sklearn.decomposition import PCA
        import matplotlib.pyplot as plt
        
        pca = PCA(pca_comps)

        red_pca = pca.fit_transform(self.red)
        red_inverted = pca.inverse_transform(red_pca)

        green_pca = pca.fit_transform(self.green)
        green_inverted = pca.inverse_transform(green_pca)

        blue_pca = pca.fit_transform(self.blue)
        blue_inverted = pca.inverse_transform(blue_pca)

        img_compressed = (np.dstack((red_inverted, green_inverted, blue_inverted))).astype(np.uint8)

        plt.imshow(img_compressed)
        plt.axis('off')
        plt.tight_layout()
        plt.savefig(self.image_name.split('.')[0]+'-compressed.jpg',bbox_inches='tight', pad_inches=0)
        
        return 'Image Compressed Successfully. Please check folder.'
```

```python
compressor = ImageCompressor()
```

```python
compressor = ImageCompressor('face_img.jpg')
```
